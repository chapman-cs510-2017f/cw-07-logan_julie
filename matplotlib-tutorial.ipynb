```python
import matplotlib.pyplot as plt
import numpy as np
```

## Matplotlib Tutorial ##

### Simple Plot ###

```python
fig, ax = plt.subplots(1,1)

X = np.linspace(-np.pi,np.pi,1000)
C, S = np.cos(X), np.sin(X)
### Plots primary functions
ax.plot(X, C, '-b', linewidth=2, label='cosine')
ax.plot(X, S, '-r', linewidth=2, label='sine')
ax.legend(loc='upper left')
### Set axis locations
ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_position(('data',0))
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

### Add labels and dashed lines
t = 2*np.pi/3
ax.plot([t, t], [np.cos(t), 0], '--b', linewidth=2)
ax.plot([t, t], [0, np.sin(t)], '--r', linewidth=2)
ax.scatter([t, t], [np.cos(t), np.sin(t)], s=50, c=['b','r'])
ax.annotate(r'$\sin\left(\frac{2\pi}{3}\right) = \frac{\sqrt{3}}{2}$', 
            xy=(t,np.sin(t)),
            xytext=(2.5,1),
            size=16,
            arrowprops=dict(arrowstyle='->', connectionstyle="arc3,rad=0.2", shrinkB=5))
ax.annotate(r'$\cos\left(\frac{2\pi}{3}\right) = -\frac{1}{2}$', 
            xy=(t,np.cos(t)),
            xytext=(0.5,-0.8),
            size=16,
            arrowprops=dict(arrowstyle='->', connectionstyle="arc3,rad=0.2", shrinkB=5))
### Set tick labels, locations, size
ax.set_xticks(np.linspace(-np.pi, np.pi, 5))
ax.set_xticklabels([r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
ax.set_yticks([-1,1])
ax.set_yticklabels([r'$-1$', r'$+1$'])
for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_size(16)
    label.set_bbox(dict(facecolor='white', edgecolor='None', alpha=0.65))
```

### Contour Plot ###

```python
fig, ax = plt.subplots()
def f(x,y):
    return (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 -y ** 2)

x = np.linspace(-3,3,100)
y = np.linspace(-3,3,100)
X, Y = np.meshgrid(x,y)

ax.contourf(X,Y, f(X,Y), 8, cmap='hot', alpha=0.75)
C = ax.contour(X,Y, f(X,Y), 8, colors='black')
ax.clabel(C, inline=1, fontsize=10)

ax.set_xticks([])
ax.set_yticks([])

ax.set_title("A Simple Contour Plot", size=16)
```

### Imshow ###

```python
ax = plt.gca()
def f(x,y):
    return (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 -y ** 2)

n = 10
x = np.linspace(-3,3, 4*n)
y = np.linspace(-3,3, 3*n)
X, Y = np.meshgrid(x,y)

plt.imshow(f(X,Y), cmap='spectral')
ax.invert_yaxis()

plt.colorbar()

ax.set_xticks([])
ax.set_yticks([])

ax.set_title("A Simple Imshow Plot", size=16)
```

### 3D Plots ###

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = Axes3D(fig)

my_cmap = 'winter'

x = np.linspace(-4,4,30)
y = np.linspace(-4,4,30)
X, Y = np.meshgrid(x,y)
Z = np.sin(np.sqrt(X**2 + Y**2))

ax.plot_surface(X,Y,Z, rstride=1, cstride=1, cmap=my_cmap)
ax.set_xlim([-4,4])
ax.set_ylim([-4,4])
ax.set_zlim([-2,2])

ax.contourf(X,Y,Z, 10, offset=-2, cmap=my_cmap)
ax.set_title("A Simple 3D Plot with Contour shadow", size=16)
```
